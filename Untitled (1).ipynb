```python

#imported all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loaded the dataset
df = pd.read_csv("f1_dnf.csv")
```

```python
#found out the number of columns and rows
df.shape
```

```python
#printed the first 5 rows 
df.head()
```

```python
#getting data info
df.info()
```

```python
#missing values
df.isnull().sum()
```

```python
df.describe()
```

```python
#filled nan values with 0
df['points'] = df['points'].fillna(0)
```

```python
# shows all the null values for a respective column
df.isnull().sum()
```

```python
#Drop all rows that have NaN in any of those columns
df = df.dropna(subset=['target_finish', 'year', 'round',])

df['year'] = pd.to_numeric(df['year'], errors='coerce')
#Convert the year into numeric data type
#if something can’t be converted,mark it as NaN.”--> error='coerce'
df['round'] = pd.to_numeric(df['round'], errors='coerce')
df['target_finish'] = pd.to_numeric(df['target_finish'], errors='coerce')  
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(x='lng', y='lat', data=df, hue='country', s=100, legend=False)
plt.title('Global Distribution of Formula 1 Circuits')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)

# tells you where (Country)the f1 racing as taken place according to longitute and latitudes 
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='surname', y='points', data=df.head(30),errorbar=None)
plt.xticks(rotation=90)
plt.title("Points by Driver")
plt.ylabel("Points")
plt.xlabel("Driver")
plt.show()

# shows points per person
# if peak points lead
#if it is not peak points are ;agging
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='grid', y='positionOrder', data=df)
plt.title("Grid Position vs Finish Position")
plt.xlabel("Starting Grid Position")
plt.ylabel("Finish Position")
plt.show()

#Starting at the front generally increases the chance of finishing the race at the top.
#Points far from the whiskers are players that are extraordinary or cpould be race failures.
```

```python
year_dnf = df.groupby('year')['target_finish'].mean()
sns.lineplot(x=year_dnf.index, y=year_dnf.values)
plt.title('Finish Rate Over the Years')
plt.ylabel('Finish Rate')
plt.xlabel('Years')
plt.show()

# If the line rises, drivers are finishing races more consistently in recent years.
#If the line drops, more drivers areNot Finishing due to various reasons
```
